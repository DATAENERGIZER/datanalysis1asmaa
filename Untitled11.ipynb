Most Selling Produst?


By Quantity

```python
data.groupby('Product')['Quantity'].sum()
```

```python
data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_Quantity=data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)  # using groupby on 2 columns and sorting the summed values.
most_Quantity
```

```python
most_Quantity=most_Quantity.reset_index()   ## turn most_quantity into a dataframe
most_Quantity
```

```python
plt.figure(figsize=(9,4))
plt.bar(most_Quantity['Product'],most_Quantity['Quantity'],color = ['red','green','blue','purple','black'] ,width = 0.4);
plt.title('Most Selling Product');
plt.xlabel('Product');
plt.ylabel('Quantity');
```
